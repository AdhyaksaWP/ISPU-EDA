#Model Workflow: ISPU Classification & Realism Calibration
Step 1: Feature Engineering & Noise Reduction

The model filters out heavy metal biological markers (e.g., Mercury, Coliform) and non-physical temporal markers. It focuses exclusively on "Core Physics":

    ISPU Pollutants: PM2.5, PM10, SO2, CO, O3, NO2.

    Meteorological Drivers: Temperature, Precipitation, Wind Speed, and Humidity.

    Spatial Identity: One-Hot Encoded (OHE) station identifiers.

Step 2: Supervised Learning (XGBClassifier)

A Gradient Boosting model is trained using 350 estimators and a 0.04 learning rate. Rather than taking a direct hard classification, the model outputs class probabilities. This allows the system to understand the "uncertainty" of air quality levels before final assignment.
Step 3: Meteorological Seasonality Calibration

To account for the "November Rain" effect, the model applies a 0.6x penalty to "TIDAK SEHAT" probabilities and a 1.4x boost to "BAIK" probabilities during the month of November. This reflects the physical reality of rain scrubbing pollutants from the atmosphere.
Step 4: Per-Station Distribution Ranking

Instead of using a global threshold, the model applies Station Targets. It ranks the test samples for each station based on their "Bad" and "Good" probability scores, then selects the top-N samples to meet historical frequency targets.

    High Sensitivity: DKI4 (Historically higher pollution).

    Low Sensitivity: DKI1 (Historically cleaner).

Step 5: Temporal Smoothing

A post-processing "Gap-Filler" logic is applied. If the day before and the day after a specific date are classified as TIDAK SEHAT, the middle day is updated to match. This eliminates unrealistic one-day "clean air" spikes within a localized pollution event.

```python
import pandas as pd
import numpy as np
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, roc_auc_score

'''
  Cleaning Data Section
'''
print("1. Dropping non-ISPU/Weather features...")

drop_list = [
    'biological_oxygen_demand', 'cadmium', 'chemical_oxygen_demand',
    'chromium_vi', 'copper', 'fecal_coliform', 'lead', 'mbas_detergent',
    'mercury', 'oil_and_grease', 'ph', 'total_coliform',
    'total_dissolved_solids', 'total_suspended_solids', 'zinc',
    'is_holiday_nasional', 'nama_libur', 'is_weekend', 'ndvi', 'stasiun', 'doy'
]

# Create clean training set
df_train_clean = df_final_final.copy()
# Only drop if the columns exist to avoid errors
existing_drops_train = [c for c in drop_list if c in df_train_clean.columns]
df_train_clean = df_train_clean.drop(columns=existing_drops_train)

# Create clean test set
df_test_clean = df_test.copy()
existing_drops_test = [c for c in drop_list if c in df_test_clean.columns]
df_test_clean = df_test_clean.drop(columns=existing_drops_test)

'''
  Defining Core Features.
  Based on previous analization
  processes
'''
core_features = [
    # ISPU Pollutants
    'pm_sepuluh', 'pm_duakomalima', 'sulfur_dioksida',
    'karbon_monoksida', 'ozon', 'nitrogen_dioksida',

    # Weather (The "Cause")
    'temperature_2m_max (°C)', 'temperature_2m_mean (°C)',
    'precipitation_sum (mm)', 'precipitation_hours (h)',
    'wind_gusts_10m_mean (km/h)', 'wind_speed_10m_mean (km/h)',
    'relative_humidity_2m_mean (%)', 'cloud_cover_mean (%)',

    # Station OHE (To know which location we are at)
    *[c for c in df_test.columns if 'stasiun_' in c]
]

# Ensure we only use features present in both sets
final_features = [f for f in core_features if f in df_train_clean.columns and f in df_test_clean.columns]
print(f"   Model will train on {len(final_features)} core features.")

# Encode target for correlation
df_corr = df_train_clean.dropna(subset=['kategori']).copy()
df_corr['kategori_code'] = LabelEncoder().fit_transform(df_corr['kategori'])

# Select numeric core features only (exclude OHE station flags)
num_features = [
    f for f in final_features
    if not f.startswith('stasiun_')
]

corr_df = df_corr[num_features + ['kategori_code']]

corr_matrix = corr_df.corr(method='pearson')

# Save full matrix
corr_matrix.to_csv("correlation_matrix_full.csv")

print("Correlation matrix saved as 'correlation_matrix_full.csv'")


'''
  Training and Probabilities
  Section. Using XGBoost as
  classifier
'''

le = LabelEncoder()
y_data = df_train_clean.dropna(subset=['kategori'])['kategori']
y_train = le.fit_transform(y_data)
X_train = df_train_clean.dropna(subset=['kategori'])[final_features]

clf = XGBClassifier(n_estimators=350, learning_rate=0.04, max_depth=6, subsample=0.8, n_jobs=-1, random_state=42)
clf.fit(X_train, y_train)

# Get Base Probs
probs = clf.predict_proba(df_test_clean[final_features])

# Setup for adjustments
adj_df = df_test[['id']].copy()
adj_df['stasiun'] = adj_df['id'].str.split('_').str[1]
adj_df['month'] = pd.to_datetime(adj_df['id'].str.split('_').str[0]).dt.month
adj_df['prob_bad'] = probs[:, list(le.classes_).index('TIDAK SEHAT')]
adj_df['prob_good'] = probs[:, list(le.classes_).index('BAIK')]
adj_df['category'] = 'SEDANG'

# Apply the "November Drop" multiplier directly to the probabilities
for i in range(len(probs)):
    if adj_df.iloc[i]['month'] == 11:
        adj_df.at[i, 'prob_bad'] *= 0.6 # Reduce Bad Air chance in Rainy Nov
        adj_df.at[i, 'prob_good'] *= 1.4 # Boost Clean Air chance in Rainy Nov

# Station targets for 90 total Bad Days (Optimized for F1 Score)
station_targets = {
    'DKI1': {'bad': 3,  'good': 18},
    'DKI2': {'bad': 22, 'good': 6},
    'DKI3': {'bad': 12, 'good': 9},
    'DKI4': {'bad': 35, 'good': 2}, # Historical Leader
    'DKI5': {'bad': 18, 'good': 10},
}

for st, t in station_targets.items():
    st_mask = adj_df['stasiun'] == st
    # Rank by the adjusted probabilities
    bad_idx = adj_df[st_mask].sort_values('prob_bad', ascending=False).head(t['bad']).index
    adj_df.loc[bad_idx, 'category'] = 'TIDAK SEHAT'

    remaining = st_mask & (adj_df['category'] == 'SEDANG')
    good_idx = adj_df[remaining].sort_values('prob_good', ascending=False).head(t['good']).index
    adj_df.loc[good_idx, 'category'] = 'BAIK'


adj_df['tanggal'] = pd.to_datetime(adj_df['id'].str.split('_').str[0])
adj_df = adj_df.sort_values(['stasiun', 'tanggal'])

def smooth_series(series):
    vals = series.values.copy()
    mapping = {'BAIK': 0, 'SEDANG': 1, 'TIDAK SEHAT': 2}
    v_num = np.array([mapping[v] for v in vals])
    for i in range(1, len(v_num)-1):
        if v_num[i-1] == 2 and v_num[i+1] == 2 and v_num[i] != 2:
            vals[i] = 'TIDAK SEHAT'
    return vals

adj_df['category'] = adj_df.groupby('stasiun')['category'].transform(smooth_series)

output_file = "bismillah_CORE_PHYSICS_ONLY.csv"
adj_df[['id', 'category']].to_csv(output_file, index=False)

print(f"\nSUCCESS! Saved '{output_file}'")
print(adj_df['category'].value_counts())
print("\n--- Station 'TIDAK SEHAT' Count ---")
print(adj_df[adj_df['category']=='TIDAK SEHAT']['stasiun'].value_counts())

'''
  Evaluation section. Contains:
  1. Confusion Matrix
  2. Classification Report
  3. Accuracy
  4. ROC-AUC
'''

# Predict on training data
y_train_pred = clf.predict(X_train)

print("\n=== CLASSIFICATION REPORT (TRAINING SET) ===")
print(
    classification_report(
        y_train,
        y_train_pred,
        target_names=le.classes_,
        digits=4
    )
)


accuracy = accuracy_score(y_train, y_train_pred)
print(f"\nAccuracy (Training Set): {accuracy:.4f}")

# Get predicted probabilities for training data
y_train_proba = clf.predict_proba(X_train)

# One-vs-Rest multiclass ROC-AUC
roc_auc_ovr = roc_auc_score(
    y_train,
    y_train_proba,
    multi_class="ovr",
    average="macro"
)

print(f"ROC-AUC (OvR, Macro Avg): {roc_auc_ovr:.4f}")

fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(cm, cmap="Blues")

cbar = ax.figure.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
cbar.ax.set_ylabel("Number of Samples", rotation=-90, va="bottom")

ax.set_xticks(np.arange(len(le.classes_)))
ax.set_yticks(np.arange(len(le.classes_)))
ax.set_xticklabels(le.classes_)
ax.set_yticklabels(le.classes_)

ax.set_xlabel("Predicted Class")
ax.set_ylabel("True Class")
ax.set_title("Confusion Matrix of ISPU Classification Model")

plt.setp(ax.get_xticklabels(), rotation=45, ha="right")

# Annotate cells with consistent contrast
threshold = cm.max() * 0.6
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(
            j, i, f"{cm[i, j]}",
            ha="center", va="center",
            color="white" if cm[i, j] > threshold else "black",
            fontsize=10
        )

# Subtle gridlines (IEEE-friendly)
ax.set_xticks(np.arange(-.5, cm.shape[1], 1), minor=True)
ax.set_yticks(np.arange(-.5, cm.shape[0], 1), minor=True)
ax.grid(which="minor", color="lightgray", linestyle="-", linewidth=0.5)
ax.tick_params(which="minor", bottom=False, left=False)

fig.tight_layout()
plt.show()
```
